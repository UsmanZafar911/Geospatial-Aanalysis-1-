```python
#Importing all relevant libraries
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
Life_Expectancy = pd.read_csv(r"C:\\Users\usman\Downloads\Geospatial\New folder\Data-20211103\API_SP.DYN.LE00.MA.IN_DS2_en_csv_v2_2825277.csv", skiprows=4)
Life_Expectancy.head()
```

```python
#Selecting two columns from the dataset(country code and 2000,2010)
WLE = Life_Expectancy[['Country Code','Indicator Name','2000', '2010']]
print(WLE)
##########Country Code Merger#############
```

```python
#Filtering the data - drop rows with missing values(NAN)
WLE.dropna()
```

```python
#Importing Naturalearth_lowers 
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
print(world.head())
```

```python
#Changing the column name
world.rename(columns = {'iso_a3':'Country Code'}, inplace = True)
print(world)
```

```python
#Merging the two datasets through common 'Country Code'.
new_df = pd.merge(world,WLE)
new_df.head() 
```

```python
new_df.plot()
```

```python
# Plot population estimates with an accurate legend - 
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
new_df.plot(column='2000', ax=ax, legend=True)
```
